```python
#11.	Replace missing values in Passenger_Rating with the mean value
df['Passenger_Rating'].fillna(df['Passenger_Rating'].mean(),inplace=True)
```

```python
#12.	Replace missing values in Meal_Option with "Unknown
df['Meal_Option'].fillna('Unknown',inplace=True)
```

```python
#13.	Replace missing values in Baggage_kg with the median value
df['Baggage_kg'] = df['Baggage_kg'].str.replace('kg', '', regex=False)
df['Baggage_kg'] = pd.to_numeric(df['Baggage_kg'], errors='coerce')
df['Baggage_kg'].fillna(df['Baggage_kg'].median(),inplace=True)
```

```python
df['Baggage_kg'].isnull().sum()
```

```python
#14.	Remove rows where Customer_ID is missing.
df.dropna(subset=['Customer_ID'],inplace=True)
```

```python
#15.	Verify that missing values are handled
df.isnull().sum()
```

```python
#16.	Remove leading and trailing spaces from all string columns
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
#17.	Convert Meal_Option values to lowercase
df['Meal_Option'] = df['Meal_Option'].str.lower()
```

```python
#18.	Convert Ticket_Type values to proper format (Refundable / Non-Refundable).
df['Ticket_Type'] = df['Ticket_Type'].str.strip().str.title()
```

```python
#19.	Standardize Checkin_Status values
df['Checkin_Status'] = df['Checkin_Status'].str.strip().str.title()
```

```python
#20.	Standardize Airport_Terminal values to T1, T2, T3 format
df['Airport_Terminal'] = df['Airport_Terminal'].str.lower().str.strip()
df['Airport_Terminal'] = df['Airport_Terminal'].replace({
    'terminal-1':'T1',
    'terminal-2':'T2',
    'terminal-3':'T3'
})
df['Airport_Terminal'] = df['Airport_Terminal'].str.upper().str.strip()
```

```python
#21.	Convert Passenger_Rating column to numeric datatype.
df['Passenger_Rating']=df['Passenger_Rating'].astype(int)
```

```python
#23.	Convert Delay_Minutes column into integer datatype.
df['Delay_Minutes'] = pd.to_numeric(df['Delay_Minutes'], errors='coerce').astype('Int64')
```

```python
df['Delay_Minutes'] = df['Delay_Minutes'].fillna(0)
```

```python
#24.	Replace "unknown" values with NaN.
import numpy as np

df['Delay_Minutes'] = df['Delay_Minutes'].replace(0, np.nan)
```
